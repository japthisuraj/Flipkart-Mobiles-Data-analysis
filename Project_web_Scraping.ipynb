```python
import requests
from bs4 import BeautifulSoup
import re
import numpy as np
import pandas as pd

# Initialize lists to store the scraped data
Brand = []
Model = []
Color = []
Storage = []
Rating = []
Ram=[]
Price= []

# Loop through the first 15 pages of Flipkart mobile listings
for page in range(1, 429):
    base_url = f"https://www.flipkart.com/mobiles/pr?sid=tyy%2C4io&otracker=categorytree&page={page}"
    request_header = {
        'Content-Type': 'text/html; charset=UTF-8',
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    }
    response = requests.get(base_url, headers=request_header)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    containers = soup.find_all("div", class_="tUxRFH")
    for container in containers:
        try:
            text = container.find("div", class_="KzDlHZ").text.strip()
            
            # Extract brand
            brand = re.findall(r"^\w+", text)
            Brand.append(brand[0] if brand else np.nan)
            
            # Extract model
            model = re.findall(r"^\w+\s([\w\s]+)\s\(", text)
            Model.append(model[0] if model else np.nan)
            
            # Extract color
            color = re.findall(r"\(([^,]+),", text)
            Color.append(color[0] if color else np.nan)
            
            # Extract storage 
            storage = re.findall(r"(\d+)\sGB",text)
            if storage:
                Storage.append(storage[0])
            else:
                Storage.append(np.nan)

            # Extract rating
            rating = container.find("div", class_="XQDdHH").text.strip() if container.find("div", class_="XQDdHH") else np.nan
            Rating.append(rating)


            
             # Extract Price
            price = container.find("div", class_="Nx9bqj _4b5DiR").text.strip() if container.find("div", class_="Nx9bqj _4b5DiR") else np.nan
            Price.append(price)
        
            
        except AttributeError:
            # Handle missing data
            continue

print("Brand:", Brand)
print("Model:", Model)
print("Color:", Color)
print("Storage:", Storage)
print("Rating:", Rating)
```

```python
# Initialize lists to store the scraped data
Screen_size =[]
Camera = []
Battery =[]
Ram=[]
# Loop through the first 15 pages of Flipkart mobile listings
for page in range(1, 429):
    base_url = f"https://www.flipkart.com/mobiles/pr?sid=tyy%2C4io&otracker=categorytree&page={page}"
    request_header = {
        'Content-Type': 'text/html; charset=UTF-8',
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    }
    response = requests.get(base_url, headers=request_header)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    containers = soup.find_all("div", class_="tUxRFH")
    for i in containers:
        #print(i.text)
        text = i.text
        ss = re.findall(r"(\d\.\d+)\sinch",text)
        if ss:
            Screen_size.append(ss[0])
        else:
            Screen_size.append(np.nan)
        

        cam = re.findall(r"Display.+\|.+Front Camera",text)
        if cam:
            Camera.append(cam[0])
        else:
            Camera.append(np.nan)

        battery = re.findall(r"(\d+)\smAh",text)
        if battery:
            Battery.append(battery[0])
        else:
            Battery.append(np.nan)

        ram = re.findall(r"(\d+)\sGB\sRAM",text)
        if ram:
            Ram.append(ram[0])
        else:
            Ram.append(np.nan)

print("Screen_size:", Screen_size)
print("Camera:", Camera)
print("Battery:", Battery)
print("Ram:", Ram)
```

```python
import pandas as pd

# Load your data
df = pd.read_csv('Flipkart.csv')

# Ensure no missing values in the 'Camera' column, fill with an empty string if necessary
df['Camera'] = df['Camera'].fillna('')

# Split the column into two new columns
split_columns = df['Camera'].str.split(r' \| ', expand=True)
df['Main Camera'] = split_columns[0].str.replace('Display', '', regex=False).str.strip()
df['Front Camera'] = split_columns[1].str.replace('Front Camera', '', regex=False).str.strip()
del df['Camera']
# Save the cleaned DataFrame to a new CSV file
df.to_csv('Flipkart.csv', index=False)

```
